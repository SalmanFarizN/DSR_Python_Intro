## Function Fundamentals

### What is a Function?

A function is a reusable block of code that performs a specific task. Functions help you:
- Avoid code repetition (DRY principle: Don't Repeat Yourself)
- Organize code into logical units
- Make code easier to test and debug
- Improve readability

### Defining and Calling Functions

### Return Values

Functions can return values using the `return` statement.

```python
# Function that returns a value
def add(a, b):
    """Add two numbers and return the result."""
    return a + b


result = add(5, 3)
print(f"5 + 3 = {result}")

# You can use the return value directly
print(f"10 + 20 = {add(10, 20)}")
```

```python
# Returning multiple values (as a tuple)
def calculate_stats(numbers):
    """Calculate min, max, and average of a list."""
    total = sum(numbers)
    count = len(numbers)
    return min(numbers), max(numbers), total / count


data = [10, 20, 30, 40, 50]
minimum, maximum, average = calculate_stats(data)

print(f"Min: {minimum}")
print(f"Max: {maximum}")
print(f"Average: {average}")
```

```python
# Function without explicit return (returns None)
def print_info(name, age):
    """Print information about a person."""
    print(f"{name} is {age} years old")
    # No return statement


result = print_info("Alice", 25)
print(f"Return value: {result}")  # None
```

## `*args` and `**kwargs`

These allow functions to accept variable numbers of arguments:
- `*args`: Captures extra positional arguments as a tuple
- `**kwargs`: Captures extra keyword arguments as a dictionary

```python
# *args example: function that sums any number of values
def sum_all(*args):
    """Sum any number of arguments."""
    print(f"Received args: {args} (type: {type(args)})")
    return sum(args)


print(sum_all(1, 2, 3))
print(sum_all(10, 20, 30, 40, 50))
```

```python
# **kwargs example: function that builds a config dict
def create_config(**kwargs):
    """Create a config dict with defaults."""
    print(f"Received kwargs: {kwargs} (type: {type(kwargs)})")
    defaults = {"debug": False, "port": 8080, "host": "localhost"}
    return {**defaults, **kwargs}  # kwargs override defaults


config = create_config(debug=True, port=3000, database="postgres")
print(f"Config: {config}")
```

```python
# Combining regular args, *args, and **kwargs
def full_example(required, *args, default="value", **kwargs):
    """Demonstrates all argument types together."""
    print(f"Required: {required}")
    print(f"Extra positional (*args): {args}")
    print(f"Default keyword: {default}")
    print(f"Extra keyword (**kwargs): {kwargs}")


full_example("must_have", 1, 2, 3, default="custom", extra="bonus")
```

```python
# Unpacking with * and **
def greet(first, last, greeting="Hello"):
    return f"{greeting}, {first} {last}!"


# Unpack a list/tuple with *
names = ["Alice", "Smith"]
print(greet(*names))  # Same as greet('Alice', 'Smith')

# Unpack a dict with **
person = {"first": "Bob", "last": "Jones", "greeting": "Hi"}
print(greet(**person))  # Same as greet(first='Bob', last='Jones', greeting='Hi')
```

## The Mutable Default Argument Trap

This is one of Python's most famous gotchas. Default arguments are evaluated once when the function is defined, not each time it's called.

```python
# BAD: Mutable default argument
def append_to_bad(item, target=[]):
    target.append(item)
    return target


print(append_to_bad(1))  # [1] - looks fine
print(append_to_bad(2))  # [1, 2] - wait, where did 1 come from?!
print(append_to_bad(3))  # [1, 2, 3] - the list persists!
```

> **Gotcha: Mutable Default Arguments**  
> Never use mutable objects (lists, dicts, sets) as default values. Use `None` and create the object inside the function.

```python
# GOOD: Use None as default, create inside function
def append_to_good(item, target=None):
    if target is None:
        target = []
    target.append(item)
    return target


print(append_to_good(1))  # [1]
print(append_to_good(2))  # [2] - fresh list each time!
print(append_to_good(3))  # [3]
```

## Lambda Functions

Lambda functions are small anonymous functions. They're useful for short operations, especially as arguments to `map()`, `filter()`, `sorted()`, and pandas methods.

```python
# Basic lambda syntax
square = lambda x: x**2
print(f"Square of 5: {square(5)}")

# Multiple arguments
add = lambda x, y: x + y
print(f"3 + 4 = {add(3, 4)}")
```

```python
# Lambdas in sorting (most common use case)
data = [{"name": "Alice", "score": 85}, {"name": "Bob", "score": 92}]
sorted_data = sorted(data, key=lambda x: x["score"], reverse=True)
print(f"Sorted by score: {sorted_data}")

# With map and filter
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Doubled: {doubled}")
print(f"Evens: {evens}")
```

> **When to use lambda vs def**  
> Use lambda for simple, one-line operations. If you need multiple lines, error handling, or a docstring, use `def`.

## Decorators

Decorators are functions that modify the behavior of other functions. They're used extensively in web frameworks (Flask routes), testing, and logging.

```python
# A decorator is a function that takes a function and returns a new function
def my_decorator(func):
    def wrapper():
        print("Before the function")
        func()
        print("After the function")

    return wrapper


# Apply decorator with @ syntax
@my_decorator
def say_hello():
    print("Hello!")


say_hello()
```

```python
# Practical example: Timing decorator
def timer(func):
    """Decorator that measures execution time."""

    @functools.wraps(func)  # Preserves function metadata
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        print(f"{func.__name__} took {end - start:.4f} seconds")
        return result

    return wrapper


@timer
def slow_function(n):
    """A deliberately slow function."""
    total = 0
    for i in range(n):
        total += i**2
    return total


result = slow_function(1_000_000)
print(f"Result: {result}")
print(f"Function name preserved: {slow_function.__name__}")
print(f"Docstring preserved: {slow_function.__doc__}")
```

```python
# Practical example: Logging decorator
def log_calls(func):
    """Decorator that logs function calls with arguments."""

    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        args_str = ", ".join(repr(a) for a in args)
        kwargs_str = ", ".join(f"{k}={v!r}" for k, v in kwargs.items())
        all_args = ", ".join(filter(None, [args_str, kwargs_str]))
        print(f"Calling {func.__name__}({all_args})")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result!r}")
        return result

    return wrapper


@log_calls
def multiply(a, b):
    return a * b


multiply(3, 4)
multiply(5, b=6)
```

```python
# Stacking decorators (applied bottom-up)
@timer
@log_calls
def compute(x, y):
    time.sleep(0.1)
    return x + y


compute(10, 20)
```

## functools.partial and functools.lru_cache

These are powerful tools from the standard library for working with functions.

```python
# functools.partial: "freeze" some arguments
def power(base, exponent):
    return base**exponent


# Create specialized versions
square = functools.partial(power, exponent=2)
cube = functools.partial(power, exponent=3)

print(f"5 squared: {square(5)}")
print(f"5 cubed: {cube(5)}")

# Useful with map/apply
numbers = [1, 2, 3, 4, 5]
cubed = list(map(cube, numbers))
print(f"All cubed: {cubed}")
```

```python
# functools.lru_cache: automatic memoization
@functools.lru_cache(maxsize=128)
def fibonacci(n):
    """Compute the nth Fibonacci number with caching."""
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


# Without caching, this would be exponentially slow
print(f"fib(40) = {fibonacci(40)}")
print(f"Cache info: {fibonacci.cache_info()}")
```

```python
# lru_cache with expensive computation
@functools.lru_cache(maxsize=32)
def expensive_query(user_id):
    """Simulate an expensive database query."""
    print(f"  [Executing query for user {user_id}...]")
    time.sleep(0.1)  # Simulate delay
    return {"user_id": user_id, "name": f"User_{user_id}"}


# First calls are slow
print("First calls:")
print(expensive_query(1))
print(expensive_query(2))

# Repeated calls are instant (cached)
print("\nCached calls:")
print(expensive_query(1))  # From cache
print(expensive_query(2))  # From cache
```

## Type Hints

Type hints make code more readable and enable IDE autocomplete and static analysis. They're increasingly important in data science codebases.

```python
# Basic type hints
def greet(name: str) -> str:
    return f"Hello, {name}!"


def add_numbers(a: int, b: int) -> int:
    return a + b


# Type hints don't enforce types at runtime
print(greet("Alice"))
print(add_numbers(3, 4))
```

```python
# Complex types using typing module
def process_items(items: List[str]) -> Dict[str, int]:
    """Count the length of each item."""
    return {item: len(item) for item in items}


def find_user(user_id: int) -> Optional[Dict[str, str]]:
    """Return user if found, None otherwise."""
    users = {1: {"name": "Alice"}, 2: {"name": "Bob"}}
    return users.get(user_id)


print(process_items(["apple", "banana", "cherry"]))
print(find_user(1))
print(find_user(999))
```

```python
# Type hints for functions as arguments (callbacks)
def apply_operation(x: int, operation: Callable[[int], int]) -> int:
    """Apply an operation to x."""
    return operation(x)


result = apply_operation(5, lambda x: x**2)
print(f"5 squared via callback: {result}")
```

```python
# Solution to Exercise 2
import time
import random


def retry(max_attempts=3):
    """
    Decorator that retries a function if it raises an exception.

    Parameters:
    -----------
    max_attempts : int
        Maximum number of attempts (default: 3)
    """

    def decorator(func):
        def wrapper(*args, **kwargs):
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    if attempt == max_attempts:
                        print(f"Failed after {max_attempts} attempts")
                        raise
                    print(f"Attempt {attempt} failed: {e}. Retrying...")
                    time.sleep(0.1)

        return wrapper

    return decorator


# Test the decorator
@retry(max_attempts=3)
def unreliable_function():
    """Function that fails randomly."""
    if random.random() < 0.7:  # 70% chance of failure
        raise ValueError("Random failure")
    return "Success!"


# Try calling it
try:
    result = unreliable_function()
    print(f"Result: {result}")
except ValueError:
    print("Function failed after all retries")
```

### Exercise 3: Memoization Decorator

Write your own `memoize` decorator that caches function results based on arguments. (This is a simplified version of `lru_cache`.)

```python
def memoize(func):
    """Cache function results based on arguments."""
    # YOUR CODE HERE
    pass


# Test:
# @memoize
# def slow_add(a, b):
#     print(f"Computing {a} + {b}...")
#     time.sleep(0.5)
#     return a + b
#
# slow_add(1, 2)  # Computes
# slow_add(1, 2)  # Returns cached result
```

```python
# Solution to Exercise 3
def memoize(func):
    """
    Decorator that caches function results.
    """
    cache = {}

    def wrapper(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]

    wrapper.cache = cache  # Expose cache for inspection
    return wrapper


# Test with Fibonacci
@memoize
def fibonacci(n):
    """Calculate nth Fibonacci number."""
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


# Test performance
import time

start = time.perf_counter()
result = fibonacci(30)
elapsed = time.perf_counter() - start

print(f"fibonacci(30) = {result}")
print(f"Time taken: {elapsed:.6f} seconds")
print(f"Cache size: {len(fibonacci.cache)} entries")
```

```python
# Solution to Exercise 4
from typing import List, Dict, Tuple


def process_user_data(
    users: List[Dict[str, any]], min_age: int = 18, sort_by: str = "age"
) -> Tuple[List[Dict[str, any]], Dict[str, float]]:
    """
    Process user data with filtering and analysis.

    Parameters:
    -----------
    users : List[Dict[str, any]]
        List of user dictionaries with 'name', 'age', 'score'
    min_age : int
        Minimum age to include (default: 18)
    sort_by : str
        Field to sort by (default: 'age')

    Returns:
    --------
    Tuple[List[Dict[str, any]], Dict[str, float]]
        Filtered and sorted users, and statistics dict
    """
    # Filter by age
    filtered = [user for user in users if user["age"] >= min_age]

    # Sort
    sorted_users = sorted(filtered, key=lambda x: x[sort_by])

    # Calculate statistics
    if filtered:
        ages = [user["age"] for user in filtered]
        scores = [user["score"] for user in filtered]
        stats = {
            "count": len(filtered),
            "avg_age": sum(ages) / len(ages),
            "avg_score": sum(scores) / len(scores),
        }
    else:
        stats = {"count": 0, "avg_age": 0, "avg_score": 0}

    return sorted_users, stats


# Test the function
test_users = [
    {"name": "Alice", "age": 25, "score": 85},
    {"name": "Bob", "age": 17, "score": 90},
    {"name": "Charlie", "age": 30, "score": 75},
    {"name": "Diana", "age": 22, "score": 95},
]

result, stats = process_user_data(test_users, min_age=18, sort_by="score")

print("Filtered and sorted users:")
for user in result:
    print(f"  {user['name']}: age {user['age']}, score {user['score']}")

print(f"\nStatistics:")
print(f"  Count: {stats['count']}")
print(f"  Avg age: {stats['avg_age']:.1f}")
print(f"  Avg score: {stats['avg_score']:.1f}")
```
