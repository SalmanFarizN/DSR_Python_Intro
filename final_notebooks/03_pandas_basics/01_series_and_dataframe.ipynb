# Pandas Series and DataFrame

Pandas is the workhorse of data analysis in Python. It provides two core data structures: Series (1D labeled array) and DataFrame (2D labeled table). Understanding these structures and their relationship is fundamental to effective data manipulation.

## Learning Objectives

- Create Series from various data sources
- Create DataFrames from dicts, lists of dicts, and NumPy arrays
- Understand the relationship between Series and DataFrame
- Use basic inspection methods: head, tail, shape, dtypes, info, describe
- Access and modify columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Pandas Series

A Series is a one-dimensional labeled array. Think of it as a column in a spreadsheet—it has values and an index (row labels).

```python
# Creating a Series from a list
s = pd.Series([10, 20, 30, 40, 50])
print("Series from list:")
print(s)
print(f"\nIndex: {s.index.tolist()}")
print(f"Values: {s.values}")
```

```python
# Creating a Series with custom index
s = pd.Series([100, 200, 300], index=['a', 'b', 'c'])
print("Series with custom index:")
print(s)

# Access by index label
print(f"\ns['b'] = {s['b']}")
```

```python
# Creating a Series from a dictionary
data = {'apples': 5, 'bananas': 3, 'oranges': 8}
s = pd.Series(data)
print("Series from dict:")
print(s)
```

```python
# Creating a Series from a scalar (broadcasts to all indices)
s = pd.Series(5, index=['a', 'b', 'c'])
print("Series from scalar:")
print(s)
```

```python
# Series operations (vectorized)
prices = pd.Series([10.5, 20.3, 15.8, 25.0], index=['A', 'B', 'C', 'D'])
quantities = pd.Series([100, 50, 75, 30], index=['A', 'B', 'C', 'D'])

# Element-wise multiplication
revenue = prices * quantities
print("Prices:")
print(prices)
print("\nQuantities:")
print(quantities)
print("\nRevenue (prices * quantities):")
print(revenue)
```

## Pandas DataFrame

A DataFrame is a two-dimensional labeled data structure—like a spreadsheet or SQL table. Each column is a Series.

```python
# Creating a DataFrame from a dict of lists
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'age': [25, 30, 35, 28],
    'city': ['Berlin', 'Munich', 'Berlin', 'Hamburg']
}

df = pd.DataFrame(data)
print("DataFrame from dict of lists:")
print(df)
```

```python
# Creating a DataFrame from a list of dicts (common when parsing JSON)
data = [
    {'name': 'Alice', 'age': 25, 'city': 'Berlin'},
    {'name': 'Bob', 'age': 30, 'city': 'Munich'},
    {'name': 'Charlie', 'age': 35, 'city': 'Berlin'},
]

df = pd.DataFrame(data)
print("DataFrame from list of dicts:")
print(df)
```

```python
# Creating a DataFrame from a NumPy array
arr = np.random.randint(0, 100, size=(4, 3))
df = pd.DataFrame(arr, columns=['A', 'B', 'C'], index=['row1', 'row2', 'row3', 'row4'])
print("DataFrame from NumPy array:")
print(df)
```

## DataFrame ↔ Series Relationship

A DataFrame is essentially a dictionary of Series sharing the same index. Each column IS a Series.

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'salary': [50000, 60000, 70000]
})

print("DataFrame:")
print(df)

# Extract a column -> returns a Series
ages = df['age']
print(f"\nType of df['age']: {type(ages)}")
print(ages)
```

```python
# Build a DataFrame from Series
names = pd.Series(['Alice', 'Bob', 'Charlie'])
ages = pd.Series([25, 30, 35])
salaries = pd.Series([50000, 60000, 70000])

df = pd.DataFrame({'name': names, 'age': ages, 'salary': salaries})
print("DataFrame built from Series:")
print(df)
```

## Basic Inspection Methods

## Column Access: df['col'] vs df.col

Both work, but bracket notation is safer and more flexible.

```python
# Bracket notation - always works
print("Using df['Age']:")
print(df['Age'].head())
```

```python
# Dot notation - convenient but risky
print("Using df.Age:")
print(df.Age.head())
```

> **Gotcha: Avoid df.col notation**  
> - Doesn't work if column name has spaces: `df.My Column` is invalid
> - Doesn't work if column name matches a DataFrame method: `df.count` returns the method!
> - Can't be used to create new columns: `df.new_col = [1,2,3]` doesn't work as expected
>
> Always use `df['column']` for safety.

```python
# Multiple columns at once -> returns a DataFrame
print("Selecting multiple columns:")
df[['Name', 'Age', 'Fare']].head()
```

## Adding and Dropping Columns

```python
# Adding a new column
df_copy = df.copy()

# From a scalar (broadcasts)
df_copy['dataset'] = 'titanic'

# From a calculation
df_copy['Fare_per_age'] = df_copy['Fare'] / df_copy['Age']

print(df_copy[['Name', 'Fare', 'Age', 'Fare_per_age', 'dataset']].head())
```

```python
# Dropping columns
# drop() doesn't modify the original by default
df_dropped = df.drop(columns=['PassengerId', 'Ticket'])
print(f"Original columns: {df.columns.tolist()}")
print(f"After drop: {df_dropped.columns.tolist()}")
```

```python
# Drop multiple columns
df_minimal = df.drop(columns=['PassengerId', 'Ticket', 'Cabin', 'Embarked'])
df_minimal.head()
```

## Visualizing Numeric Column Distributions

```python
# Get numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numeric columns: {numeric_cols}")

# Plot distributions
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

for i, col in enumerate(numeric_cols[:6]):
    ax = axes[i]
    df[col].hist(bins=30, ax=ax, color='steelblue', edgecolor='black', alpha=0.7)
    ax.set_title(col)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    
    # Add statistics
    mean = df[col].mean()
    median = df[col].median()
    ax.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.1f}')
    ax.axvline(median, color='green', linestyle='-', label=f'Median: {median:.1f}')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```
