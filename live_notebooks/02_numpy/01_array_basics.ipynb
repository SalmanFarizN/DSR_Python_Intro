# NumPy Array Basics

NumPy is the foundation of numerical computing in Python. While Python lists are flexible, they're slow for mathematical operations because each element is a full Python object. NumPy arrays store data in contiguous memory blocks with a fixed type, enabling vectorized operations that are orders of magnitude faster.

## Why NumPy?

Python lists are versatile but slow for numerical work:
- Each element is a full Python object with overhead
- Operations require Python loops (slow)
- Not optimized for math operations

NumPy arrays:
- Store data in contiguous memory (cache-friendly)
- Fixed data type = no per-element overhead
- Vectorized operations implemented in C (fast!)
